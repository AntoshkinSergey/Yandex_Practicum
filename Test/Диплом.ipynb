### Блок пользовательских функций

```python
def mf_bar_labels(axis, orient='v', space=0.08, total=0, prec=0):
    """ Проставляет значения столбчатых диаграмм, orient - ориентация диаграммы (v, h)
        total - для построения долей рядом с абсолютными значениями
        space - сколько отступать от столбца в % от размера столбца
    """
    
    # рассчитывается точность отображения данных
    # prec = '{}' if prec== 0 else '{:.'+str(prec)+'f}'
    prec = '{:.'+str(prec)+'f}'
    
    for i, rect in enumerate(axis.patches):
        # рассчет высоты - для вертикальной, ширины - для горизонтальной гистограммы
        dist = rect.get_height() if orient == 'v' else rect.get_width()
          
        if total == 0: # выводить только абсолютные значения
            val = prec.format(dist)
        else: # выводить с долями к total
            val = prec.format(dist) + ' ('+ str(round(dist*100/total,1)) +'%)'
        
        if orient == 'v':
            axis.text(   
                i # bar index (x - координаты текста)
                , dist*(1+space) # (y - координаты текста)
                , val # y label - текст для вывода
                , ha='center' 
                , va='center'
            )
        else:
            axis.text(                      
                dist*(1+space) # x coordinate of text - координаты текста
                , i # bar index (y coordinate of text)
                , val  # x label - текст для вывода
                , ha='center' 
                , va='center' 
            )
```

### Исследовательский анализ данных, рассчеты и добавление резульатов

```python
# рассмотрим модель формирования журнала событий игроков
# в разрезе столбцов event, building_type, project_type
(
    df_game_actions
    .pivot_table(index=['event', 'building_type','project_type'], aggfunc={'user_id':'count'})
    .rename(columns={'user_id':'количество записей в журнале'})
)
```

```python
# исследуем прогресс игры пользователей по количеству комплексных состояний: событий-построек-проектов
user_progress = df_game_actions.copy() # для анализа создаю копию датасета

# формирем столбец состояний
user_progress['combined'] =(
    user_progress['event']+' * '+user_progress['building_type']+' * '+ user_progress['project_type']
)

# по каждому пользователю оставляем уникальные состояния
user_progress2 = user_progress[['user_id', 'combined']].drop_duplicates().reset_index(drop=True)
user_progress2
```

```python
# по каждому пользователю считаем кол-во уникальных состояний
users = user_progress2.pivot_table(index='user_id', aggfunc='count').reset_index()

# находим количество пользователей по количеству состояний
users2=users.pivot_table(index='combined', aggfunc='count')

#выводим в табличном виде информацию
display(users2)

# визуализируем график (по-быстрому без красот, для себя - чисто технический анализ, не для презентации)
users2.plot.bar(
    xlabel='Кол-во уникальных состояний игрока'
    , ylabel='Кол-во игроков'
    , title='Количество игроков по количеству уникальных действий'
    , grid=True
);
```

#### Определение когорты пользователя

```python
# когорту пользователя определим по первому событию, выполненному в игре, но предаварительно проверим гипотизу о том, 
# что день клика пользователя по рекламе совпадает с днем появления первого события игрока

# по каждому пользователю найдем самое раннее (для определения индекса когорты) 
# и самое позднее событие (для вычисления времени, проведенное пользователем в игре)
df_user_event_dt = (
    df_game_actions.pivot_table(index='user_id', values='event_datetime', aggfunc=['min', 'max'])
    .droplevel(1, axis=1) # удаляем лишний уровень в названии столбцов
    .rename(columns={'min':'first_event_dt', 'max':'last_event_dt'}) # переименовываем столбцы
)

# дату-время переведем в дату и возьмем значение даты в качестве идентификатора когорты
df_user_event_dt['cohort'] = df_user_event_dt.first_event_dt.dt.date 

# посмотрим когорты по количеству игроков
print('Когорты по количеству игроков:\n', df_user_event_dt.cohort.value_counts())
```

* в дальнейшем для анализа стоимости рекламных кампаний по источникам трафика и когортам примем допущение, что дата когорты стоимости рекламы на день отстает от даты игровой когорты,

#### Определение времени в игре, проведенного пользователями

За время, проведенное пользователями в игре будет взят интервал по часам от первого до последнего события.

```python
# находим разницу между первым и последним событием игрока в часах
df_user_event_dt['play_time'] = (df_user_event_dt.last_event_dt - df_user_event_dt.first_event_dt)/pd.Timedelta(1,'h')

# округление игрового времени до 1 знака после запятой
df_user_event_dt.play_time = round(df_user_event_dt.play_time,1)

# выводим первые 5 строк полученной таблицы
df_user_event_dt.head()
```

#### Рассчет основной таблицы исследования

```python
# формируем основную таблицу исследования
df_main = (
    df_user_event_dt[['cohort', 'play_time']] # для всех пользователей определяем когорту и кол-во часов в игре
    .join(df_user_source.set_index('user_id')) # присоединям информацию об источнике трафика
    .join(  # присоединям информацию о количестве построенных зданий игроками
        ( # подсчет построенных зданий игроками
            df_game_actions[df_game_actions.event=='building']
            .pivot_table(index='user_id', aggfunc={'event':'count'})
            .rename(columns={'event':'build_count'})
        ))
    .join(# игроки, выполнившие проект
        (
            df_game_actions[df_game_actions.event=='project'][['user_id']]
            .set_index('user_id', drop=False).rename(columns={'user_id':'project'})
        ))
    .join(# игроки, завершившие 1 уровень
        (
            df_game_actions[df_game_actions.event=='finished_stage_1'][['user_id']]
            .set_index('user_id', drop=False).rename(columns={'user_id':'finished'})
        ))
    .reset_index()
)

# преобразуем столбцы project и finished в булев тип
df_main[['project', 'finished']] = df_main[['project', 'finished']].notnull()

# смотрим структуру и первые 5 строк новой таблицы
df_main.info()
df_main.head()
```

### Проверка статистических гипотез
**Методические пояснения**
Прежде чем проверять статистическую гипотезу необходимо:
* определиться с выбором показателя (случайной величины), по значениям которого будут проверены стат.гипотизы,
* выдвинуть нулевую гипотезу и альтернативную,
* исследовать характер распределения показателя с целью обоснования выбора стат критерия.

#### Cравнение времени завершения уровня в зависимости от способа прохождения (с и без проекта)
Рассматриваемая случайная величина - время прохождения 1 уровня, в подготовленном датасете df_main это столбец "play_time", значения которого выражены в часах.
Получим две выборки:
* группа А (датасет df_a) - значения времени прохождения 1 уровня игроками, без выполненного проекта,
* группа В (датасет df_b) - значения времени прохождения 1 уровня игроками, c выполненным проектом.

Для пользователя целью игры является завершение 1 уровня, при этом выполнение проекта (сборка орбитального спутника) является необязательной опцией, выполнение которой требует дополнительного времени для его реализации. В этой связи предположим, что среднее время прохождения 1 уровня у игроков с реализованным проектом должно быть больше среднего времени прохождения игроков без проекта.

За нулевую гипотезу (Н0) примем утверждение, что среднее время группы А совпадает с показателем по группе В. За альтернативную гипотизу (Н1) - среднее время группы А меньше, чем у группы В.

Выберем порог уровня статистической значимости 5% (alpha = 0.05).

```python
# подготовим 2 выборки А и В
df_a = df_main.query("finished==True and project==False")[['play_time']]
df_b = df_main.query("finished==True and project==True")[['play_time']]
```

```python
# построю таблицу описательной статистики по двум выборкам
df_a.describe().join(df_b.describe(), lsuffix='_A', rsuffix='_B')
```

**Промежуточные выводы**:
* средние групп А и В отличаются и составляют 267 и 323 часа соответственно, что в целом подтверждает наше предположение о том, что среднее время прохождения 1 уровня больше у угроков с проектом по сравлению игроков без проекта,
* стандартные отклонения показателя отличаются незначительно, т.е. разброс значений случайной величины группы А чуть больше (98 часов), чем у группы В (84 часа).

```python
# посмотрим на распределение показателя по группам
list1={'группы А':df_a, 'группы В':df_b} # задаем словарь групп с их данными
for gr in list1: # цикл по группам
    plt.figure(figsize=(15,5))
    ax1 = plt.subplot(1,2,1)
    ax1 = sns.distplot(list1[gr]['play_time'], ax=ax1) # построение плотности распределения
    ax1.set_title(f"График плотности вероятности {gr}")
    ax1.set_xlabel("Время прохождения, часы")
    ax1.set_ylabel("Относительная частота")
    
    ax2 = plt.subplot(1,2,2)
    st.probplot(list1[gr]['play_time'], dist="norm", plot=ax2) # проверка на нормальность распределения по qq-графикам 
    plt.show()
```

В этой связи предположим, что среднее время прохождения 1 уровня у игроков с реализованным проектом должно быть больше среднего времени прохождения игроков без проекта.

```python
# проведем стат тест
alpha = 0.05
results = st.mannwhitneyu(df_a, df_b, alternative='less')
print(f"p-значение: {round(results.pvalue[0]*100,1)}%")

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
